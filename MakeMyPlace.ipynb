```python
#Installed web scraping libraries or packages
#!pip install lxml html5lib beautifulsoup4
import requests
import pandas as pd
```

```python
#Data Extraction from HTML using pandas library or package
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
#Read the contents from the website
dfs = pd.read_html(url)
#Three tables on this website because the output of following command is 3
#print(len(dfs))
#Extracting first table or required table or dfs[0]
#print(dfs[0])
TorontoTable = []
TorontoTable = dfs[0]
TorontoTable.head(15)
```

```python
# Data Cleaning: Drop "Not Assigned" Borough values
TorontoTable = TorontoTable[TorontoTable.Borough != 'Not assigned']
TorontoTable.head()
# Filling missing data values: As suggested replace "Not Assigned" Neighbourhood with Borough value
TorontoTable['Neighbourhood'] = TorontoTable.apply(lambda x: x['Borough'] if x['Neighbourhood']=='Not assigned' else x['Neighbourhood'], axis=1)
TorontoTable.head(15)
```
